## Cleaning of data for better understanding and Summurization more further.

```python
df["Killed"]=df["Killed"].fillna(0)
df["Wound"]=df["Wound"].fillna(0)
df["Casualty"]=df["Killed"]+df["Wound"]
```

```python
df.describe()
```

```python
attacks=df["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head()
```

## Data Visualization:
### 1. Year vs Year

```python
attacks.plot(kind="bar",color="skyblue",figsize=(17,7),fontsize=14)
plt.title("Time1ine of Attacks", fontsize=16)
plt.xlabel("Years" , fontsize=16)
plt.ylabel("Number of Attacks" , fontsize=16)
plt.show()
```

## 2. Region vs Attacks with diff. parameters:

```python
yt=df[["Year","Casualty"]].groupby("Year").sum()
yt.head()
```

```python
yt.plot(kind="bar",color="skyblue",figsize=(15,6))
plt.title("Year wise Casualties",fontsize=13)
```

```python
yk=df[["Year","Killed"]].groupby("Year").sum()
yk.head()
```

```python
yw=df[["Year","Wound"]].groupby("Year").sum()
yw.head()
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,1,1)
ax1=fig.add_subplot(2,1,2)

#Killed
yk.plot(kind="bar",color="skyblue",figsize=(16,16),ax = ax0)
ax0.set_title("People Killed in each Year")
ax0.set_xlabel("years")
ax0.set_ylabel("Number of People Killed")

#Wound
yw.plot(kind="bar",color="skyblue",figsize=(15,15),ax=ax1)
ax1.set_title("People Wounded in each Year")
ax1.set_xlabel("Years")
ax1.set_ylabel("Number of People Wound")

plt.show()
```

```python
reg= pd.crosstab(df.Year, df.Region)
reg.head()
```

```python
reg.plot(kind="area", stacked=False, alpha=0.5, figsize=(20,10))
plt.title("Region wise attacks", fontsize = 20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(15,6))
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
rc=df[["Region","Casualty"]].groupby("Region").sum().sort_values(by="Casualty",ascending=False)
rc
```

```python
rc.plot(kind="bar", color="skyblue",figsize=(15,6))
plt.title("Region wise Casulaties", fontsize=13)
plt.xlabel("Regions", fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties", fontsize=13)
plt.show()
```

```python
rk=df[["Region","Killed"]].groupby("Region").sum().sort_values(by="Killed",ascending=False)
rk
```

```python
rw=df[["Region","Wound"]].groupby("Region").sum().sort_values(by="Wound",ascending=False)
rw
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

#Killed
rk.plot(kind="bar", color="skyblue", figsize=(15,6),ax=ax0)
ax0.set_title("People Killed in each Region")
ax0.set_xlabel("Regions")
ax0.set_ylabel("Number of People Wounded")

#Wound
rw.plot(kind="bar", color="skyblue", figsize=(15,6),ax=ax0)
ax1.set_title("People Killed in each Region")
ax1.set_xlabel("Regions")
ax1.set_ylabel("Number of People Wounded")

plt.show()
```
